```python
"""""1. Make a calculator using Python with addition , subtraction , 
multiplication ,division and power."""""
  operation = input("Select an operator:n")
  print("Enter two numbers")
  number_1 = int(input())
  number_2 = int(input())

  if operation == '+': # To add two numbers
    print("sum=",number_1 + number_2)

  elif operation == '-': # To subtract two numbers
    print("subtraction:",number_1 - number_2)

  elif operation == '*': # To multiply two numbers
    print("Multiplication:",number_1 * number_2)

  elif operation == '/': # To divide two numbers
    print("division:",number_1 / number_2)

  else:
    print('Invalid Input')
```

```python
"""""5. Write a program to identify duplicate values from list."""""

lst = [ 3, 6, 9, 12, 3, 30, 15, 9, 45, 36, 12, 12]
dupItems = []
uniqItems = {}
for x in lst:
   if x not in uniqItems:
      uniqItems[x] = 1
   else:
      if uniqItems[x] == 1:
         dupItems.append(x)
      uniqItems[x] += 1
print(dupItems)
```
